```python
education = pd.read_csv("/content/drive/MyDrive/Data Open Championship 2022/Cleaned dataset/Final_Education_Fairness.csv")
```

```python
gini = education.loc[:,['year','Gini_gender', 'Gini_ethnics','Gini_income']]
```

```python
gini = gini.melt('year', var_name = 'Gini_index', value_name='Value')
```

```python
gini.head()
```

```python
plt.figure(figsize=(17,5))
sns.lineplot(x="year", y="Value", hue='Gini_index', data=gini).set(title='Gini Indexes for Public Institutions from 2015-2020')
plt.legend(bbox_to_anchor=(1.01, 0.65),loc='upper left', title='Gini Index', borderaxespad=0)
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x='year', y='edu_fairness_score',data=education).set(title='Education Fairness Score from 2015 - 2020')
```
