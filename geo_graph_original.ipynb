```python
nomi = pgeocode.Nominatim('us')
result = nomi.query_postal_code("07601")
result["latitude"], result["longitude"]
```

```python
df_uni["lat"] = pd.Series()
df_uni["lon"] = pd.Series()
```

```python
# generate the latitude and longitude
# reference to https://pypi.org/project/pgeocode/

for idx, uni in df_uni.iterrows():
    if np.isnan(uni["lat"]):
        if idx % 100 == 0:
            print(idx)
        zipcode = uni["zip"][:5]
        result = nomi.query_postal_code(zipcode)
        df_uni.at[idx, "lat"] = result["latitude"]
        df_uni.at[idx, "lon"] = result["longitude"]
```
