```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
```

```python
race = pd.read_csv('population-of-the-us-by-race-2000-2021.csv')
education_equity = pd.read_csv('Education_Equity.csv')
education_equity = education_equity.iloc[: , 1:]
education_equity.head()
```

```python
race.head()
```

```python
race['Total'] = race['White'] + race['Black'] + race['American Indian'] + race['Asian'] + race['Native Hawaiian'] + race['2 Or More Races'] + race['Hispanic']
```

```python
race['%White_us'] = race['White'] / race['Total']
race['%Black_us'] = race['Black'] / race['Total']
race['%American Indian_us'] = race['American Indian'] / race['Total']
race['%Asian_us'] = race['Asian'] / race['Total']
race['%Native Hawaiian_us'] = race['Native Hawaiian'] / race['Total']
race['%2 Or More Races_us'] = race['2 Or More Races'] / race['Total']
race['%Hispanic_us'] = race['Hispanic'] / race['Total']
```

```python
race = race.drop(columns = ['Total','White', 'Black', 'American Indian', 'Asian', 'Native Hawaiian', '2 Or More Races', 'Hispanic'])
```
